```python
import math
import pandas as pd

from random import randint
from numpy.random import choice
from tkinter import *
from tkinter import ttk

#creo un vector de ubers y solicitudes para almacenar a los datos y poder trabajarlos de manera más fácil    
ubers = []
requests = []
estadistica = []

#Se declaran las variables que almacenarán el algoritmo, la jornada y la solicitud a evaluar
ventana = Tk()    #variable contenedora de la ventana que contiene el menu
algoritmo = StringVar()    #variable que almacenará el algoritmo seleccionado por el usuario en el menu
jornada = StringVar()    #variable que almacenará la jornada seleccionado por el usuario en el menu
solicitud = StringVar()    #variable que almacenará la solicitud seleccionado por el usuario en el menu
Combo_algoritmos = ttk.Combobox(ventana)    #instancia de combobox que almacenará los algoritmos a seleccionar 
Combo_jornada = ttk.Combobox(ventana)    #instancia de combobox que almacenará las jornadas a seleccionar

Combo_request = ttk.Combobox(ventana)    #instancia de combobox que almacenará las solicitudes a seleccionar
```

```python
'''
Nombre: importar_jornada
tipo: Funcion
Entrada: nombre del archivo de entrada a leer y que contiene la jornada a evaluar (jornada).
Proceso: Se define que columnas tendrá el archivo una vez cargado en memoria (jornada_cols). El archivo referido por la entrada
se almacena en un panda (pd).
Salida: Se retorna el documento con las jornadas cargadas a partir del archivo especificado en la entrada (df_jornada).
'''

def importar_jornada(jornada):
    #Importo el archivo de jornadas, defino las columnas que tendrá el archivo y lo almaceno en un DataFrame
    jornada_cols = ['posx', 'posy']
    df_jornada = pd.read_csv(jornada, header = None, sep=' ', names = jornada_cols)
    return df_jornada


'''
Nombre: importar_solicitudes
tipo: Funcion
#Entrada: nombre del archivo de entrada a leer y que contiene las solicitudes a evaluar (solicitud).
#Proceso: Se define que columnas tendrá el archivo una vez cargado en memoria (request_cols). El archivo referido por la entrada
se almacena en un panda (pd). Luego este archivo se ordena de menor a mayor, es decir, desde las primeras solicitudes hasta las 
últimas, en intervalos de tiempo.
#Salida: Se retorna el documento con las solicitudes cargadas a partir del archivo especificado en la entrada (df_request).
'''

def importar_solicitudes(solicitud):
    #Importo el archivo de solicitudes, defino las columnas que tendrá el archivo y lo almaceno en un DataFrame
    request_cols = ['posx_i', 'posy_i', 'sep1', 'posx_f', 'posy_f', 'sep2', 'tiempo']
    df_request = pd.read_csv(solicitud, header = None, sep=' ', names = request_cols)
    
    #Ordeno el archivo en orden ascendente, desde las solicitudes más próximas a las más lejanas
    df_request = df_request.sort_values(['tiempo'], ascending=True)
        
    return df_request
```

```python
'''
Nombre: Uber
Tipo: Clase
Descripción: Clase que permite almacenar a los uber que daran solución al problema de los k-servers. Se almacena su posicion
en coordenadas x (posx) e y (posy), asi como el tiempo en el que éste está disponible para atender solicitudes (time) y la 
cantidad que éste fue utilizado por los clientes, vale decir, clientes atendidos (uses). La clase está equipada con un 
constructor y los respectivos métodos para agregar(set_) y consultar datos (get_).
'''

class Uber:
    posx = ""
    posy = ""
    time = ""
    uses = ""
    
    def __init__(self, x, y):
        self.posx = x
        self.posy = y
        self.time = 0
        self.uses = 0
    
    def get_posx(self):
        print (self.posx)
    
    def get_posy(self):
        print (self.posy)
    
    def get_time(self):
        print (self.time)
    
    def set_posx(self, new_posx):
        self.posx = new_posx
        
    def set_posy(self, new_posy):
        self.posy = new_posy
    
    def set_time(self, new_time):
        self.time = new_time 
```

```python
'''
Nombre: Request
Tipo: Clase
Descripción: Clase que permite almacenar a las solicitudes de los clientes. Se almacena su posicion inicial en coordenadas x 
(posx_ini) e y (posy_ini) y las coordenadas de destino (posx_fin, posy_fin). Se almacena tambien el tiempo en el que éste 
desea ser atentido (time) y un estado final (Disponible/Available "A", Atendido/Closed "C"). La clase está equipada con un constructor y los 
respectivos métodos para agregar(set_) y consultar datos (get_).
'''

class Request:
    posx_ini = ""
    posy_ini = ""
    posx_fin = ""
    posy_fin = ""
    time = ""
    estado = ""
        
    def __init__(self, a, b, c, d, e):
        self.posx_ini = a
        self.posy_ini = b
        self.posx_fin = c
        self.posy_fin = d
        self.time = e
        self.estado = "A"
            
    def get_posx_ini(self):
        print (self.posx_ini)
    
    def get_posy_ini(self):
        print (self.posy_ini)
        
    def get_posx_fin(self):
        print (self.posx_fin)
    
    def get_posy_fin(self):
        print (self.posy_fin)    
    
    def get_time(self):
        self.time = new_time
    
    def set_posx_ini(self, new_posx):
        self.posx_ini = new_posx
     
    def set_posy_ini(self, new_posy):
        self.posy_ini = new_posy
        
    def set_posx_fin(self, new_posx):
        self.posx_fin = new_posx
    
    def set_posy_fin(self, new_posy):
        self.posy_fin = new_posy    
    
    def set_time(self, new_time):
        self.time = new_time
```

```python
'''
Nombre: Candidatos_Harmonic
Tipo: Clase
Descripción: Clase que permite almacenar a los uber disponibles para atender las solicitudes de clientes, por medio del algoritmo
Armónico. En esta clase se almacena la posición de los candidatos para resolver la consulta, dentro del arreglo de Ubers . 
Se almacena tambien la distancia desde el uber al cliente, la probabilidad definida por la formula del algoritmo armonico y el
normalizador o ponderador. La clase está equipada con un constructor y los respectivos métodos para agregar(set_) y consultar
datos (get_).
'''

class Candidatos_Harmonic:
    pos_arreglo = ""
    dist_objetivo = ""
    probabilidad = ""
    ponderador = ""
            
    def __init__(self, a, b, c):
        self.pos_arreglo = a
        self.dist_objetivo = b
        self.probabilidad = c
        self.ponderador = 0
        
        
    def get_pos(self):
        print (self.pos_arreglo)
    
    def get_dist(self):
        print (self.dist_objetivo)
        
    def get_probabilidad(self):
        print (self.probabilidad)
    
    def get_ponderador(self):
        print (self.ponderador)
        
    def set_pos(self, new_pos):
        self.pos_arreglo = new_pos
    
    def set_dist(self, new_dist):
        self.dist_objetivo = new_dist    
    
    def set_probabilidad(self, new_prob):
        self.probabilidad = new_prob
    
    def set_ponderador(self, new_pond):
        self.ponderador = new_pond
        
```

```python
'''
Nombre: add_uber
tipo: Funcion
Entrada: vector que almacenará objetos de la clase Uber.
Proceso: Cada uno de los uber obtenidos del documento jornada (Uber), se almacenan en un vector(uber_list).
Salida: -
'''
def add_uber(uber_list, Uber):
    uber_list.append(Uber)

    
'''
Nombre: add_request
tipo: Funcion
Entrada: vector que almacenará objetos de la clase Request.
Proceso: Cada una de las solicitudes obtenidos del documento solicitudes (Request), se almacenan en un vector(request_list).
Salida: -
'''
def add_request(request_list, Request):
    request_list.append(Request)
    

'''
Nombre: add_candidatos
tipo: Funcion
Entrada: vector que almacenará objetos de la clase Candidatos_Harmonic.
Proceso: Cada uno de los candidatos para atender a los clientes según el algoritmo armónico, se almacenan en un vector llamado
un vector(candidatos_list).
Salida: -
'''
def add_candidatos(candidatos_list, Candidatos_Harmonic):
    candidatos_list.append(Candidatos_Harmonic)
    

'''
Nombre: calcular_distancia
tipo: Funcion
Entrada: coordenadas de dos puntos (x1,y1) y (x2,y2)
Proceso: Esta función calcula la distancia entre dos puntos utilizando la fórmula de la distancia de Manhattan.
Salida: Distancia entre dos puntos (distancia)
'''    
def calcular_distancia(posx1, posy1, posx2, posy2):
    distancia = math.sqrt(math.pow(posx1 - posx2, 2) + math.pow(posy1 - posy2, 2))
    return distancia


'''
Nombre: search_4Uber_Greedy
tipo: Funcion
Entrada: posición del cliente (posx_ini, posy_ini), número de uber consultado (nr_request), cantidad total de ubers (nro_ubers).
Proceso: Se verifica que exista un uber capaz de atender la solicitud del cliente. Para lo anterior, se revisa si el uber está
disponible, en función del tiempo en el que éste se desocupará. Cuando se encuentra un uber disponible para atender la demanda,
este se almacena en un arreglo. Luego, de todos los potenciales uber capaces de resolver la solicitud, se calcula la distancia
a la que se encuentran del cliente y se selecciona el que esté más cerca. En caso de que no exista ningún uber disponible en 
el momento en que se solicitó, el cliente tendrá que esperar hasta que alguno se desocupe.
Salida: posición del uber seleccionado (posicion), dentro del arreglo en el que están almaenados los uber.
'''   
def search_4Uber_Greedy(posx_ini, posy_ini, nr_request, nro_ubers):
    posicion = 0
    contador = 0
    menor_distancia = calcular_distancia(0, 0, 1000, 1000)
    contador_candidatos = 0
    cercano = ubers[0].time
    candidatos = []
        
    while(contador_candidatos==0):
        for i in range(nro_ubers):
            if ubers[i].time <= requests[nr_request].time:
                candidatos.append(contador)
                contador_candidatos+=1
                distancia = calcular_distancia(ubers[i].posx, ubers[i].posy, posx_ini, posy_ini)
                if menor_distancia > distancia:
                    menor_distancia = distancia
                    posicion = contador
            else:
                if(cercano >= ubers[i].time):
                    cercano = ubers[i].time
            contador+=1
        if(contador_candidatos==0):
            contador = 0
            requests[nr_request].time += cercano
    
    return posicion



'''
Nombre: search_4Uber_Random
tipo: Funcion
Entrada: Cantidad de ubers durante la jornada y el número de uber consultado (nr_request).
Proceso: Se verifica que exista un uber capaz de atender la solicitud del cliente. Para lo anterior, se revisa si el uber está
disponible, en función del tiempo en el que éste se desocupará. Cuando se encuentra un uber disponible para atender la demanda,
este se almacena en un arreglo. Luego, de todos los potenciales uber capaces de resolver la solicitud, se selecciona uno en forma 
aleatoria. En caso de que no exista ningún uber disponible en el momento en que se solicitó, el cliente tendrá que esperar hasta
que alguno se desocupe.
Salida: posición del uber seleccionado (seleccionado), dentro del arreglo en el que están almaenados los uber.
'''   
def search_4Uber_Random(nro_ubers, nr_request):
    posicion = 0
    seleccionado = 0
    contador = 0
    contador_candidatos = 0
    cercano = ubers[0].time
    candidatos = []
    
    while(contador_candidatos==0):
        for i in range(nro_ubers):
            if ubers[i].time <= requests[nr_request].time:
                candidatos.append(contador)
                contador_candidatos+=1
            else:
                if(cercano >= ubers[i].time):
                    cercano = ubers[i].time
            contador+=1
        if(contador_candidatos==0):
            contador = 0
            requests[nr_request].time += cercano
    inicio = 0
    fin = len(candidatos)-1
    posicion = randint(inicio,fin)
    seleccionado = candidatos[(posicion)]
    return seleccionado


'''
Nombre: search_4Uber_Harmonic
tipo: Funcion
Entrada: posición del cliente (posx_ini, posy_ini), número de uber consultado (nr_request), cantidad total de ubers (nro_ubers).
Proceso: Se verifica que exista un uber capaz de atender la solicitud del cliente. Para lo anterior, se revisa si el uber está
disponible, en función del tiempo en el que éste se desocupará. Cuando se encuentra un uber disponible para atender la demanda,
este se almacena en un arreglo. Luego, de todos los potenciales uber capaces de resolver la solicitud, se calcula la distancia
a la que se encuentran del cliente y, utilizando la formula del algoritmo armónico, se calcula un valor ponderado de probabilidad
para cada potencial uber, cuyo valor va de 0 a 1 y que actua como un "peso" que dirigirá, en cierta forma, la selección del
uber que resolvera la solicitud. La selección se realiza en forma aleatoria, pero utilizando los pesos ya mencionados. En caso 
de que no exista ningún uber disponible en el momento en que se solicitó, el cliente tendrá que esperar hasta que alguno se 
desocupe.
Salida: posición del uber seleccionado (posicion), dentro del arreglo en el que están almaenados los uber.
'''     
def search_4Uber_Harmonic(posx_ini, posy_ini, nr_request, nro_ubers):
    posicion = 0
    contador = 0
    contador_candidatos = 0
    cercano = ubers[0].time
    normalizador = 0
    candidatos = []
    probab = []
    pos_matriz = []
    
    while(contador_candidatos==0):
        for i in range(nro_ubers):
            if ubers[i].time <= requests[nr_request].time:
                distancia = calcular_distancia(ubers[i].posx, ubers[i].posy, posx_ini, posy_ini)
                if(distancia == 0):
                    distancia = 1
                    add_candidatos(candidatos, Candidatos_Harmonic(contador, distancia, 1))
                else:
                    add_candidatos(candidatos, Candidatos_Harmonic(contador, distancia, 1/distancia))
                normalizador += (1/distancia)
                contador_candidatos+=1
            else:
                if(cercano >= ubers[i].time):
                    cercano = ubers[i].time
            contador+=1
        if(contador_candidatos==0):
            contador = 0
            requests[nr_request].time += cercano
    
    for i in range(contador_candidatos):
        candidatos[i].ponderador = (candidatos[i].probabilidad)/normalizador
        pos_matriz.append(candidatos[i].pos_arreglo)
        probab.append(candidatos[i].ponderador) 
    
    posicion = choice(pos_matriz, p=probab)
    
    return posicion



'''
Nombre: carga_datos_ubers
tipo: Funcion
Entrada: DataFrame con los uber cargados en un panda (df_jornada), según la jornada.
Proceso: Se toman los datos recogidos del archivo de entrada y se almacenan en un vector de objetos (ubers).
Salida: -
'''   
def carga_datos_ubers(df_jornada):
    for i in range(len(df_jornada)):
        add_uber(ubers, Uber(df_jornada.loc[i,'posx'],df_jornada.loc[i,'posy']))

        
'''
Nombre: carga_datos_solicitudes
tipo: Funcion
Entrada: DataFrame con las solicitudes cargados en un panda (df_request).
Proceso: Se toman los datos recogidos del archivo de entrada y se almacenan en un vector de objetos (requests).
Salida: -
'''   
def carga_datos_solicitudes(df_request):
    for i in range(len(df_request)):
        add_request(requests, Request(df_request.loc[i,'posx_i'], df_request.loc[i,'posy_i'], df_request.loc[i,'posx_f'], df_request.loc[i,'posy_f'], df_request.loc[i,'tiempo']))

        
'''
Nombre: inicializa_vector_est
tipo: Funcion
Entrada: vector donde se almacenarán los costos por intervalo de tiempo (estadistica) y la opción que se quiere que se ejecute(opcion).
Proceso: Se toma el vector de la entrada y se inicializa con valores 0.
Salida: -
'''
def inicializa_vector_est(estadistica, opcion):
    if(opcion == 0):
        for i in range(12001):
            estadistica.append(0)
    else:
        estadistica.clear()


'''
Nombre: inicializa_est_req
tipo: Funcion
Entrada: Dataframe con las solicitudes cargadas.
Proceso: Se toma el vector de las solicitudes y se inicializa el estado del cliente con disponible (Available).
Salida: -
'''
def inicializa_est_req(df_request):
    for i in range(len(df_request)):
        requests[i].estado = "A"



'''
Nombre: nro_autos
tipo: Funcion
Entrada: posición del uber seleccionado (posicion), la distancia que recorrera en el viaje completo (distancia) y el vector de
costos (estadistica).
Proceso: Se toma el vector de costos, se situa el cabezal en el intervalo de tiempo en el que se acepta un viaje y se agrega 
la cantidad de autos que están funcionando en cada intervalo de tiempo.
Salida: -
'''
def nro_autos(posicion, distancia, estadistica):
    dist = int(round(distancia))
    pos = int(round(posicion))
    for i in range(dist):
        if(((pos+1+i) + dist) > 12000):
            i = dist
        else:
            estadistica[(pos+1) + i] = estadistica[(pos+1) + i] + 1

            
            
'''
Nombre: funcion_costos
tipo: Funcion
Entrada: vector de costos (estadistica).
Proceso: Se toma el vector de costos, se situa el cabezal al inicio y se comienza a acumular la distancia recorrida por todos los
uber en cada intervalo de tiempo.
Salida: -
'''
def funcion_costos(estadistica):
    for i in range(len(estadistica)):
        if(i==0):
            estadistica[i] = estadistica[i]
        else:
            estadistica[i] += estadistica[i-1]
    
    
'''
Nombre: algoritmo_greedy
tipo: Funcion
Entrada: Dataframe con la información sobre los uber y las solicitudes a responder.
Proceso: Se recorren todas las solicitudes una por una buscando un uber capaz de responder dicha solicitud en el menor tiempo 
posible. Para lo anterior, se busca al Uber más cercano y se asigna al cliente. Luego se cambia el estado del cliente a "closed"
y la posición final en la que quedo el uber, asi como la distancia recorrida y el número de veces que ha sido llamado.
Salida: -
'''
#Buscar solicitudes sin atender, con Algoritmo Greedy
def algoritmo_greedy(df_request, df_jornada):
    for i in range(len(df_request)):
        if requests[i].estado == "A":
            uber_libre = search_4Uber_Greedy(requests[i].posx_ini, requests[i].posy_ini, i, len(df_jornada))
            ubers[uber_libre].time += calcular_distancia(ubers[uber_libre].posx, ubers[uber_libre].posy, requests[i].posx_ini, requests[i].posy_ini) + calcular_distancia(requests[i].posx_ini, requests[i].posy_ini, requests[i].posx_fin, requests[i].posy_fin)
            nro_autos(requests[i].time, calcular_distancia(ubers[uber_libre].posx, ubers[uber_libre].posy, requests[i].posx_ini, requests[i].posy_ini) + calcular_distancia(requests[i].posx_ini, requests[i].posy_ini, requests[i].posx_fin, requests[i].posy_fin), estadistica)
            ubers[uber_libre].posx = requests[i].posx_fin
            ubers[uber_libre].posy = requests[i].posy_fin
            ubers[uber_libre].uses += 1
            requests[i].estado = "C"
    funcion_costos(estadistica)

    
'''
Nombre: algoritmo_random
tipo: Funcion
Entrada: Dataframe con la información sobre los uber y las solicitudes a responder.
Proceso: Se recorren todas las solicitudes una por una buscando todos los uber capaces de responder dicha solicitud seleccionando
al azar a quien terminará por avanzar hacia el cliente. Luego se cambia el estado del cliente a "closed"
y la posición final en la que quedo el uber, asi como la distancia recorrida y el número de veces que ha sido llamado.
Salida: -
'''
#Buscar solicitudes sin atender, con Algoritmo Random
def algoritmo_random(df_request, df_jornada):
    for i in range(len(df_request)):
        if requests[i].estado == "A":
            uber_libre = search_4Uber_Random(len(df_jornada), i)
            ubers[uber_libre].time = ubers[uber_libre].time + calcular_distancia(ubers[uber_libre].posx, ubers[uber_libre].posy, requests[i].posx_ini, requests[i].posy_ini) + calcular_distancia(requests[i].posx_ini, requests[i].posy_ini, requests[i].posx_fin, requests[i].posy_fin)
            nro_autos(requests[i].time, calcular_distancia(ubers[uber_libre].posx, ubers[uber_libre].posy, requests[i].posx_ini, requests[i].posy_ini) + calcular_distancia(requests[i].posx_ini, requests[i].posy_ini, requests[i].posx_fin, requests[i].posy_fin), estadistica)
            ubers[uber_libre].posx = requests[i].posx_fin
            ubers[uber_libre].posy = requests[i].posy_fin
            ubers[uber_libre].uses += 1
            requests[i].estado = "C"
    funcion_costos(estadistica)

    
'''
Nombre: algoritmo_armonico
tipo: Funcion
Entrada: Dataframe con la información sobre los uber y las solicitudes a responder.
Proceso: Se recorren todas las solicitudes una por una buscando todos los uber capaces de responder dicha solicitud. Se calcula
la distancia entre todos los posibles candidatos y el cliente que hizo la solicitud. Se realiza un calculo probabilistico
para cada uno de estos candidatos con respecto a la función armónica y se le asigna un peso a cada uno de estos, el cual
dirigirá la seleccion del uber, selección que se hace al azar. Luego se cambia el estado del cliente a "closed" y la posición 
final en la que quedo el uber, asi como la distancia recorrida y el número de veces que ha sido llamado.
Salida: -
'''    
#Buscar solicitudes sin atender, con Algoritmo Armónico
def algoritmo_armonico(df_request, df_jornada):
    for i in range(len(df_request)):
        if requests[i].estado == "A":
            uber_libre = search_4Uber_Harmonic(requests[i].posx_ini, requests[i].posy_ini, i, len(df_jornada))
            ubers[uber_libre].time = ubers[uber_libre].time + calcular_distancia(ubers[uber_libre].posx, ubers[uber_libre].posy, requests[i].posx_ini, requests[i].posy_ini) + calcular_distancia(requests[i].posx_ini, requests[i].posy_ini, requests[i].posx_fin, requests[i].posy_fin)
            nro_autos(requests[i].time, calcular_distancia(ubers[uber_libre].posx, ubers[uber_libre].posy, requests[i].posx_ini, requests[i].posy_ini) + calcular_distancia(requests[i].posx_ini, requests[i].posy_ini, requests[i].posx_fin, requests[i].posy_fin), estadistica)
            ubers[uber_libre].posx = requests[i].posx_fin
            ubers[uber_libre].posy = requests[i].posy_fin
            ubers[uber_libre].uses += 1
            requests[i].estado = "C"
    funcion_costos(estadistica)

    
'''
Nombre: imprime
tipo: Funcion
Entrada: Dataframe con la información sobre los uber
Proceso: Se imprime la posición final de los uber, luego de haber atendido todas las solicitudes, el camino total recorrido y
la cantidad de veces que fueron contactados y utilizados por clienes.
Salida: -
'''    
def imprime(df_jornada):
    #Imprimir resultados obtenidos despues de ejecutar un algoritmo
    suma = 0
    for i in range(len(df_jornada)):
        print(f"{ubers[i].posx} {ubers[i].posy} {ubers[i].time} {ubers[i].uses}")
        suma += ubers[i].time

        
'''
Nombre: exporta_datos
tipo: Funcion
Entrada: Algoritmo seleccionado, jornada, solicitudes, vector de costos, DataFrame de las solicitudes.
Proceso: Se genera la estructura del nombre del archivo de salida y la extensión. El nombre del archivo es:
Algoritmo-Jornada-Solicitud.dat y dentro contiene la información recopilada por el vector de costos.
Salida: -
'''
def exporta_datos(algoritmo,jornada, solicitud, estadistica, df_request):
    archivo = open("{0}-{1}-{2}".format(algoritmo,jornada, solicitud), "w")
    for i in range(len(estadistica)):
        archivo.write("{0} {1}\n".format(i,estadistica[i]))
    archivo.close()
    inicializa_vector_est(estadistica,1)
    inicializa_est_req(df_request)
    
```
