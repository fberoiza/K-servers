CONCLUSIONES

Es fácil notar que aquel algoritmo que tuvo un mejor rendimiento en cuanto al costo total recorrido, fue el algoritmo Greedy, para todas sus variantes, sean estas, cantidad de autos y clientes. En segundo lugar se encuentra el algoritmo Armónico y finalmente el algoritmo Random. No es dificil entender que una estrategia de selección al azar, siempre será la peor decisión y, en consecuencia, aquella que demuestre un peor desempeño, pues no tiene una heuristica que ayude en la toma de decisiones, por tanto decide sin ningún criterio pudiendo en muchos casos, seleccionar la peor decision de todas las posibles, lo que a final de cuentas impactará en el costo total de la solución. 

Por otra parte, el algoritmo greedy funciona totalmente diferente, pues de todas las decisiones posibles, siempre elige la mejor. Bajo este punto de vista, Greedy tiene una heuristica que le permite dirimir sobre cual es la mejor opción, por tanto, realiza una busqueda informada y esa es una mejor técnica para problemas de este tipo.

Finalmente el algoritmo Armónico viene a ser una mezcla entre Greedy y Random pues, entre todas las opciones, sabe quien es mejor, sin embargo no selecciona la mejor, asigna una probabilidad de elección, pero termina eligiendo en forma aleatoria. En rigor, no es una busqueda aleatoria pura, pues está dirigida por la probabilidad de elección seleccionada, de manera que existe más probabilidades de elegir una mejor opción, pero no es una probabilidad absoluta, pues siempre puede elegir una opcion medianamente buena o la peor opción.

Por todo lo anterior es que los resultados se repitieron en todos los escenarios. Algoritmo greedy siempre fue el más eficiente, seguido por Harmonic y, en ultimo lugar, siempre fue Random. En resumen, una estrategia de búsqueda informada, aunque la lógica no sea muy buena o cercana al patrón de comportamiento, siempre será mejor que elegir al azar.
